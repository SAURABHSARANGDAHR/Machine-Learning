

# Matplotlib Pyplot

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4])
plt.xlabel('X label')
plt.ylabel('Y label')
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16], 'ys')
plt.axis([0,6,0,20])
plt.show()
```

```python
import numpy as np
t=np.arange(0.,5.,0.2)
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.show()
```
